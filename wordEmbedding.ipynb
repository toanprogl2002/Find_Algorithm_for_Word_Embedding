```python
import pandas as pd
import re
from nltk import ngrams
from gensim.models import Word2Vec
from importlib import reload
import sys
if sys.version[0] == '2':
    reload(sys)
    sys.setdefaultencoding("utf-8")
```

```python

pd.options.display.max_rows = 9999

df = pd.read_csv("TruyenKieu_data.txt",sep='/', names=["row"], encoding="utf8")
df.head(10)
```

```python
def transform_row(row):
    row = re.sub(r"^[0-9\.]+","",row)
    row = re.sub(r"[\.,\?]+$","", row)
    row = row.replace(",", " ").replace(".", " ")\
        .replace(";"," ").replace( "“", " ")\
        .replace(":", " ").replace("”", " ") \
        .replace('"', " ").replace("'", " ") \
        .replace("!", " ").replace("?", " ") \
        .replace("  ", " ")
    row = row.strip()
    return row
df["row"] = df.row.apply(transform_row)
df.to_csv("pre_text.txt", index= False, header=None)
```

```python
from nltk.util import ngrams
def kieu_gram(strList, n):
    n_gram = list(ngrams(strList.split(),n))
    return [" ".join(gram).lower() for gram in n_gram]
df["1gram"] = df.row.apply(lambda t: kieu_gram(t,1))
df["2gram"] = df.row.apply(lambda t: kieu_gram(t,2))
```

```python
df["context"] = df["1gram"] + df["2gram"]
```

```python
df
```

```python
train_data = df.context.tolist()
```

```python
import logging


model = Word2Vec(sentences = train_data, vector_size=100, window=5, min_count=3, workers=4)
```

```python
model.save("word2vec_truyenKieu.model")
```
